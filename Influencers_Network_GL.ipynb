```python
import pandas as pd
import requests
import time
```

```python
def scrape_links(frame, base_url_to_search, topic):
    r = requests.get(base_url_to_search)
    tag = r.json()

    end_cursor = tag["graphql"]["hashtag"]["edge_hashtag_to_media"]["page_info"]["end_cursor"]
    has_next_page = tag["graphql"]["hashtag"]["edge_hashtag_to_media"]["page_info"]["has_next_page"]
    result = frame.copy()
    for i in range(len(tag["graphql"]["hashtag"]['edge_hashtag_to_media']["edges"])):
        df_temp = pd.io.json.json_normalize(tag["graphql"]["hashtag"]['edge_hashtag_to_media']["edges"][i])
        short_code = df_temp["node.shortcode"][0]
        url = "https://www.instagram.com/p/%s/?__a=1"%short_code
        df_temp["username"] = requests.get(url).json()["graphql"]["shortcode_media"]["owner"]["username"]
        result = pd.concat([result,df_temp],axis=0)
        result["topic"] = topic

    return result, has_next_page, end_cursor
```
